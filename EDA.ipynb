```python
total_runs = pd.DataFrame(data.groupby(['batting_team','bowling_team','match_id','striker'])['runs_off_bat'].sum()).\
       rename(columns={"runs_off_bat": "Batsman_runs"})

#number of sixes
sixes = pd.DataFrame(data[data['runs_off_bat'] == 6].groupby(['batting_team','bowling_team','match_id', 'striker'])['runs_off_bat'].count()).\
            rename(columns={"runs_off_bat": "6"})

#number of fours
fours = pd.DataFrame(data[data['runs_off_bat'] == 4].groupby(['batting_team','bowling_team','match_id', 'striker'])['runs_off_bat'].count()).\
            rename(columns={"runs_off_bat": "4"})

#no of balls
balls_faced = pd.DataFrame(data.groupby(['batting_team','bowling_team','match_id', 'striker'])['ball'].nunique()).\
            rename(columns={"ball": "total_legal_balls_faced"})

#strikerate
Strikerate = pd.DataFrame((data.groupby(['batting_team','bowling_team','match_id','striker'])['runs_off_bat'].sum()/data.groupby(['batting_team','bowling_team','match_id', 'striker'])['ball'].nunique())*100).\
       rename(columns={"runs_off_bat":"strike_rate"})


#fifties
fifties = pd.DataFrame((data.groupby(['batting_team','bowling_team','match_id','striker'])['runs_off_bat'].sum() >= 50)  
                       ).\
       rename(columns={"runs_off_bat": "50"})
#hundreds
hundreds = pd.DataFrame(data.groupby(['batting_team','bowling_team','match_id','striker'])['runs_off_bat'].sum() >= 100  ).\
       rename(columns={"runs_off_bat": "100"})

#duck
duck  = pd.DataFrame(data.groupby(['batting_team','bowling_team','match_id','striker'])['runs_off_bat'].sum() == 0).\
        rename(columns={"runs_off_bat": "duck"})

# #batsmen_position
# batsmen_position = pd.DataFrame(df_2021.groupby(['battingteam','bowlingteam','matchid', 'batsmanname'])['fallofwickets'].min())

#batting_team
# batting_team = pd.DataFrame(df_2021['battingteam'])

# batting_team
# bowling_team = pd.DataFrame(df_2021['bowlingteam'])
```

```python
import os
import pandas as pd

# Specify the folder containing the CSV files
folder = "path/to/folder"

# Initialize an empty list to store the dataframes
df_list = []

# Iterate through all files in the folder
for file in os.listdir(folder):
  # Only consider CSV files
  if file.endswith(".csv"):
    # Read in the CSV file
    df = pd.read_csv(os.path.join(folder, file))
    # Add the dataframe to the list
    df_list.append(df)

# Concatenate all dataframes into a single dataframe
df = pd.concat(df_list)

# Write the combined data to a new CSV file
df.to_csv("combined_file.csv", index=False)
```
