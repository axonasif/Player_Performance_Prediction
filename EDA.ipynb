```python
import os
import pandas as pd

# Specify the input and output folders
input_folder = "/workspace/Player_Performance_Prediction/Demo"
output_folder = "/workspace/Player_Performance_Prediction/CHECK"

# Iterate through all files in the input folder
for file in os.listdir(input_folder):
    # Only consider CSV files
    if file.endswith(".csv"):
        # Read in the CSV file
        df = pd.read_csv(os.path.join(input_folder, file))

        # Perform some operations on the dataframe, such as filtering rows or calculating values
        data = df.copy()
        data = data.fillna(0)

        total_runs = pd.DataFrame(data.groupby(['batting_team', 'bowling_team', 'match_id', 'striker'])['runs_off_bat'].sum())
        total_runs = total_runs.rename(columns={"runs_off_bat": "Batsman_runs"})

        # Number of sixes
        sixes = pd.DataFrame(data[data['runs_off_bat'] == 6].groupby(['batting_team', 'bowling_team', 'match_id', 'striker'])['runs_off_bat'].count())
        sixes = sixes.rename(columns={"runs_off_bat": "6"})

        # Number of fours
        fours = pd.DataFrame(data[data['runs_off_bat'] == 4].groupby(['batting_team', 'bowling_team', 'match_id', 'striker'])['runs_off_bat'].count())
        fours = fours.rename(columns={"runs_off_bat": "4"})

        # Number of balls
        balls_faced = pd.DataFrame(data.groupby(['batting_team', 'bowling_team', 'match_id', 'striker'])['ball'].nunique())
        balls_faced = balls_faced.rename(columns={"ball": "total_legal_balls_faced"})

        # Strike rate
        strikerate = pd.DataFrame((data.groupby(['batting_team', 'bowling_team', 'match_id', 'striker'])['runs_off_bat'].sum() / data.groupby(['batting_team', 'bowling_team', 'match_id', 'striker'])['ball'].nunique()) * 100)
        strikerate = strikerate.rename(columns={"runs_off_bat": "strike_rate"})

        # Fifties
        fifties = pd.DataFrame(data.groupby(['batting_team', 'bowling_team', 'match_id', 'striker'])['runs_off_bat'].sum() >= 50)
        fifties = fifties.rename(columns={"runs_off_bat": "50"})

        # Hundreds
        hundreds = pd.DataFrame(data.groupby(['batting_team', 'bowling_team', 'match_id', 'striker'])['runs_off_bat'].sum() >= 100)
        hundreds = hundreds.rename(columns={"runs_off_bat": "100"})

        # Duck
        duck = pd.DataFrame(data.groupby(['batting_team', 'bowling_team', 'match_id', 'striker'])['runs_off_bat'].sum() == 0)
        duck = duck.rename(columns={"runs_off_bat": "duck"})

        total_runs['Balls'] = balls_faced
        total_runs['Fours'] = fours
        total_runs['Sixes'] = sixes
        # total_runs['SR'] = sr['0']
        total_runs['Fifties'] = fifties
        total_runs['Hundreds'] = hundreds
        total_runs['Duck'] = duck

        # total_runs = total_runs.fillna(0)
        # Fill missing values with 0
        total_runs = total_runs.fillna(0)

        # Make a copy of the dataframe
        df1 = total_runs.copy()

        # Replace boolean values with 1 or 0
        df1['Hundreds'] = df1['Hundreds'].replace({True: 1, False: 0})
        df1['Fifties'] = df1['Fifties'].replace({True: 1, False: 0})
        df1['Duck'] = df1['Duck'].replace({True: 1, False: 0})
        df1['Sixes'] = df1['Sixes'].replace({True: 1, False: 0})

        # Save the modified dataframe to a CSV file
        df1.to_csv("df.csv")

        # Read the CSV file into a new dataframe
        sample = pd.read_csv("/workspace/Player_Performance_Prediction/df.csv")

        # Calculate the dreamll score
        dreamll_score = sample['Batsman_runs'] + sample['Sixes']*2 + sample['Fours'] + sample['Fifties']*8 + sample['Hundreds']*16 - sample['Duck']*2 + 4

        # Add the dreamll score to the dataframe
        sample['dreamll_score'] = dreamll_score

        # Save the resulting dataframe to a CSV file in the output folder
        sample.to_csv(os.path.join(output_folder, file))

```
